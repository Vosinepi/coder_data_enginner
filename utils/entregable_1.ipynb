```python
import requests as req
import sqlalchemy as sql
import pandas as pd
import datetime as dt


from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook, curdoc
from bokeh.models import ColumnDataSource, HoverTool, LabelSet, Label
from bokeh.models.annotations import Label
from bokeh.layouts import column, row, gridplot

output_notebook()
```

### api cryptos

```python
def precio_historico(symbol, exchange='binance', after='2020-01-01'):
    url = "https://api.cryptowat.ch/markets/{exchage}/{symbol}usdc/ohlc".format(exchage=exchange, symbol=symbol)
    response = req.get(url, params={
        'periods': '3600', 
        'after': str(int(pd.Timestamp(after).timestamp()))})
    response.raise_for_status()
    data = response.json()
    df = pd.DataFrame(data['result']['3600'], columns = ['CloseTime', 'OpenPrice', 'HighPrice', 'LowPrice', 'ClosePrice', 'Volume', 'NA'])
    df['NombreMoneda'] = symbol
    df['CloseTime'] = pd.to_datetime(df['CloseTime'], unit='s')
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
ultimos_365_dias = (dt.datetime.now() - pd.offsets.Day(365))
periodo = ultimos_365_dias
# Convert datetime object to string
after_date = periodo.strftime('%Y-%m-%d')

btc = precio_historico('btc', 'kucoin', after=after_date)
eth = precio_historico('eth', 'kucoin', after=after_date)
op = precio_historico('op', 'kucoin', after=after_date)
bnb = precio_historico('bnb', 'kucoin', after=after_date)
ada = precio_historico('ada', 'kucoin', after=after_date)
doge = precio_historico('doge', 'kucoin', after=after_date)
dot = precio_historico('dot', 'kucoin', after=after_date)
xrp = precio_historico('xrp', 'kucoin', after=after_date)
ltc = precio_historico('ltc', 'kucoin', after=after_date)
link = precio_historico('link', 'kucoin', after=after_date)
bch = precio_historico('bch', 'kucoin', after=after_date)
```

```python
fondo = '#202124'


TOOLTIPS = [
    ('date', '@CloseTime{%Y-%m-%d %H:%M:%S}'),
    ("Precio", "@ClosePrice{0.0}"),
    ("Volume", "@Volume{(0.00 a)}"),
]

curdoc().theme = 'dark_minimal'

p1 = figure(x_axis_type='datetime', title="btc", width=1000, tooltips=TOOLTIPS)
p1.grid.grid_line_alpha = 0.3
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Price'
p1.line(color='blue', legend_label='Btc', x='CloseTime', y='ClosePrice', source=btc)
p1.legend.location = "top_left"
label_btc = Label(x=btc['CloseTime'].iloc[-1], y=btc['ClosePrice'].iloc[-1], text=str(btc['ClosePrice'].iloc[-1]), text_color='white')
p1.add_layout(label_btc)

p2 = figure(x_axis_type='datetime', title="eth", width=1000, tooltips=TOOLTIPS)
p2.grid.grid_line_alpha = 0.3
p2.xaxis.axis_label = 'Date'
p2.yaxis.axis_label = 'Price'
p2.line(color='green', legend_label='Eth', x='CloseTime', y='ClosePrice', source=eth)
p2.legend.location = "top_left"
label_eth = Label(x=eth['CloseTime'].iloc[-1], y=eth['ClosePrice'].iloc[-1], text=str(eth['ClosePrice'].iloc[-1]), text_color='white')
p2.add_layout(label_eth)

p3 = figure(x_axis_type='datetime', title='op', width=1000, tooltips=TOOLTIPS)
p3.grid.grid_line_alpha = 0.3
p3.xaxis.axis_label = 'Date'
p3.yaxis.axis_label = 'Price'
p3.line(color='red', legend_label='Op', x='CloseTime', y='ClosePrice', source=op)
p3.legend.location = "top_left"
label_op = Label(x=op['CloseTime'].iloc[-1], y=op['ClosePrice'].iloc[-1], text=str(op['ClosePrice'].iloc[-1]), text_color='white')
p3.add_layout(label_op)

p4 = figure(x_axis_type='datetime', title='bnb', width=1000, tooltips=TOOLTIPS)
p4.grid.grid_line_alpha = 0.3
p4.xaxis.axis_label = 'Date'
p4.yaxis.axis_label = 'Price'
p4.line(color='orange', legend_label='Bnb', x='CloseTime', y='ClosePrice', source=bnb)
p4.legend.location = "top_left"
label_ada = Label(x=bnb['CloseTime'].iloc[-1], y=bnb['ClosePrice'].iloc[-1], text=str(bnb['ClosePrice'].iloc[-1]), text_color='white')
p4.add_layout(label_ada)

# make a grid
grid = gridplot([p1, p2, p3, p4], ncols=2, width=500, height=400)

show(grid)
```
